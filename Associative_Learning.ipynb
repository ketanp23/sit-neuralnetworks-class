```python
import numpy as np

# Patterns to associate (binary, -1/1 for bipolar)
patterns = np.array([[1, 1, -1], [ -1, 1, 1]])  # Two 3-bit patterns

# Hebbian learning: Compute weight matrix
W = np.dot(patterns.T, patterns) / patterns.shape[1]
np.fill_diagonal(W, 0)  # No self-connections

# Noisy input (flip one bit of first pattern)
input_pattern = np.array([1, -1, -1])

# Recall: Update until stable
max_iter = 10
for _ in range(max_iter):
    output = np.sign(np.dot(W, input_pattern))
    if np.all(output == input_pattern):
        break
    input_pattern = output

print("Recalled pattern:", output)  # Should recall [1, 1, -1]
```
