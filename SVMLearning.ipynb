```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay

# Data (use first two features for viz)
iris = datasets.load_iris()
X = iris.data[:, :2]  # Sepal length/width
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Linear SVM
linear_svm = SVC(kernel='linear', C=1.0)
linear_svm.fit(X_train, y_train)
linear_pred = linear_svm.predict(X_test)
print("Linear SVM Accuracy:", accuracy_score(y_test, linear_pred))

# RBF SVM
rbf_svm = SVC(kernel='rbf', gamma=0.5, C=1.0)
rbf_svm.fit(X_train, y_train)
rbf_pred = rbf_svm.predict(X_test)
print("RBF SVM Accuracy:", accuracy_score(y_test, rbf_pred))

# Visualize RBF boundary (simplified)
disp = DecisionBoundaryDisplay.from_estimator(rbf_svm, X, response_method="predict", cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors="k")
plt.title("RBF SVM on Iris")
plt.show()
```
